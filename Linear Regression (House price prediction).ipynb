```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("E:\\Real_Estate.csv")
df.head()
```

# Simple Linear Regression

```python
plt.figure(figsize = (10,6))
plt.scatter(df['Distance to the nearest MRT station'] , df['House price of unit area'])
plt.xlabel('Distance to the nearest MRT station')
plt.ylabel('House price of unit area')
plt.title('Plot')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
X = df['Distance to the nearest MRT station'].values.reshape(-1,1)
y = df['House price of unit area'].values
```

```python
from sklearn.linear_model import LinearRegression
model  = LinearRegression()
model.fit(X , y)
```

```python
y_pred = model.predict(X)
```

```python
plt.figure(figsize = (10,6))
plt.scatter(df['Distance to the nearest MRT station'] , df['House price of unit area'])
plt.plot(df['Distance to the nearest MRT station'] ,y_pred  , color= "red")
plt.xlabel('Distance to the nearest MRT station')
plt.ylabel('House price of unit area')
plt.title('Plot')
plt.legend()
plt.grid(True)
plt.show()
```

# Multiple Linear Regression

```python
df2 = pd.read_csv("E:\\Real_Estate.csv")
df2.head()
```

```python
plt.figure(figsize = (10,6))
plt.scatter(df2['Distance to the nearest MRT station'] , df2['House price of unit area'] , df2['House age'])
plt.xlabel('Distance to the nearest MRT station')
plt.ylabel('House price of unit area')
plt.title('Plot')
plt.legend()
plt.grid(True)
plt.show()
```

```python
X = df2[['Distance to the nearest MRT station', 'House age']]
y = df2['House price of unit area']
```

```python
from sklearn.linear_model import LinearRegression
model1  = LinearRegression()
model1.fit(X , y)
```

```python
y_pred2 = model1.predict(X)
```

```python
plt.figure(figsize = (10,6))
plt.scatter(df2['Distance to the nearest MRT station'] , df2['House price of unit area'] , df2['House age'])
plt.plot(df2['Distance to the nearest MRT station'] ,y_pred2  , color= "red")
plt.xlabel('Distance to the nearest MRT station')
plt.ylabel('House price of unit area')
plt.title('Plot')
plt.legend()
plt.grid(True)
plt.show()
```
